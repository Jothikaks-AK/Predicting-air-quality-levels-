```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1️⃣ Load dataset
df = pd.read_csv('/content/global_air_quality_data_10000 (1).csv')
print("Original Data:\n", df)

# 2️⃣ Handle missing values (fill with mean)
df['PM2.5'].fillna(df['PM2.5'].mean(), inplace=True)
df['PM10'].fillna(df['PM10'].mean(), inplace=True)
df['NO2'].fillna(df['NO2'].mean(), inplace=True)
print("\nAfter Handling Missing Values:\n", df)

# 3️⃣ Remove duplicates
df.drop_duplicates(inplace=True)
print("\nAfter Removing Duplicates:\n", df)

# 4️⃣ Outlier detection and removal (IQR method)
Q1 = df[['PM2.5', 'PM10', 'NO2']].quantile(0.25)
Q3 = df[['PM2.5', 'PM10', 'NO2']].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[['PM2.5', 'PM10', 'NO2']] < (Q1 - 1.5 * IQR)) | (df[['PM2.5', 'PM10', 'NO2']] > (Q3 + 1.5 * IQR))).any(axis=1)]
print("\nAfter Removing Outliers:\n", df)

# 5️⃣ Encode categorical variables
df = pd.get_dummies(df, columns=['City'])
print("\nAfter Encoding Categorical Variables:\n", df)

# 6️⃣ Feature scaling
scaler = StandardScaler()
features_to_scale = ['PM2.5', 'PM10', 'NO2']
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
print("\nAfter Scaling:\n", df)

# 7️⃣ Train-test split
X = df.drop(columns=[])  # No explicit target here; update with your real target column if available
y = None  # Replace with your target variable if available
# Since we don't have a target in this sample, we'll just show how to split features
X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)
print("\nTrain Set:\n", X_train)
print("\nTest Set:\n", X_test)
```
